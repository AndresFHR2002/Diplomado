## Creaci√≥n de dimensiones

```python
# Dimension de tiempo
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_tiempo;")

# Crear la tabla usando el campo 'a_o'
query_dim_tiempo = """
CREATE TABLE dim_tiempo AS
SELECT 
    ROW_NUMBER() OVER () AS id_tiempo,
    a_o AS anio
FROM educacion_raw
WHERE a_o IS NOT NULL
GROUP BY a_o
ORDER BY a_o;
"""

conn.execute(query_dim_tiempo)
```

```python
# DIMENSI√ìN MUNICIPIO (solo con df_divipola)
dim_municipio = df_divipola[[
    "cod_mpio", "nom_mpio", "cod_dpto", "dpto", "tipo_municipio", "longitud", "latitud"
]].drop_duplicates()

dim_municipio = dim_municipio.rename(columns={
    "cod_mpio": "id_municipio",
    "nom_mpio": "municipio",
    "cod_dpto": "id_departamento",
    "dpto": "departamento"
})

dim_municipio.head(15)
```

```python
# Dimension de municipio
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_municipio;")

# Crear tabla con los nombres correctos desde la API
query_dim_municipio = """
CREATE TABLE dim_municipio AS
SELECT 
    ROW_NUMBER() OVER () AS id_municipio,
    e.c_digo_municipio AS codigo_municipio,
    e.municipio,
    d.dpto AS nombre_departamento,
    d.tipo_municipio,
    CAST(d.longitud AS FLOAT) AS longitud,
    CAST(d.latitud AS FLOAT) AS latitud
FROM educacion_raw e
LEFT JOIN divipola d 
    ON e.c_digo_municipio = d.cod_mpio
WHERE e.c_digo_municipio IS NOT NULL AND e.municipio IS NOT NULL
GROUP BY 
    e.c_digo_municipio, e.municipio, d.dpto, d.tipo_municipio, d.longitud, d.latitud;
"""

conn.execute(query_dim_municipio)
```

```python
# Dimension de Departamento
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_departamento;")

# Crear la tabla usando nombres corregidos y √∫nicos
query_dim_departamento = """
CREATE TABLE dim_departamento AS
SELECT 
    ROW_NUMBER() OVER () AS id_departamento,
    c_digo_departamento AS codigo_departamento,
    departamento
FROM educacion_raw
WHERE c_digo_departamento IS NOT NULL AND departamento IS NOT NULL
GROUP BY c_digo_departamento, departamento
ORDER BY departamento;
"""

conn.execute(query_dim_departamento)
```

## Tabla de hechos

```python
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS hechos_educacion_limpia;")

# Crear tabla de hechos con control de duplicados y datos limpios
query_hechos_limpia = """
CREATE TABLE hechos_educacion_limpia AS
SELECT 
    t.id_tiempo,
    m.id_municipio,
    d.id_departamento,
    AVG(CASE 
        WHEN e.tasa_matriculaci_n_5_16 > 100 THEN 100
        WHEN e.tasa_matriculaci_n_5_16 < 0 THEN 0
        ELSE e.tasa_matriculaci_n_5_16
    END) AS tasa_matriculacion,
    AVG(e.cobertura_neta) AS cobertura_neta
FROM educacion_raw e
JOIN dim_tiempo t 
    ON e.a_o = t.anio
JOIN dim_municipio m 
    ON e.c_digo_municipio = m.codigo_municipio AND e.municipio = m.municipio
JOIN dim_departamento d 
    ON e.c_digo_departamento = d.codigo_departamento AND e.departamento = d.departamento
WHERE 
    e.tasa_matriculaci_n_5_16 IS NOT NULL AND
    e.cobertura_neta IS NOT NULL
GROUP BY 
    t.id_tiempo, m.id_municipio, d.id_departamento;
"""

conn.execute(query_hechos_limpia)
```

## 2. ¬øC√≥mo comparar√≠a el rendimiento educativo por municipios?

```python
query_2 = """
SELECT 
    m.municipio,
    ROUND(AVG(e.aprobaci_n), 2) AS promedio_aprobacion,
    ROUND(AVG(e.reprobaci_n), 2) AS promedio_reprobacion,
    ROUND(AVG(e.deserci_n), 2) AS promedio_desercion
FROM educacion_raw e
JOIN dim_municipio m ON e.c_digo_municipio = m.codigo_municipio
GROUP BY m.municipio
ORDER BY promedio_aprobacion DESC
LIMIT 10;
"""

df_p2 = pd.read_sql(query_2, conn)
df_p2
```

## 3. ¬øQue departamentos son los que mejor cobertura tienen? ¬øPueden hacer c√°lculo con SQL?

```python
query_cobertura_ponderada_fix = """
SELECT 
    d.departamento,
    ROUND(
        SUM(e.cobertura_neta * e.poblaci_n_5_16) / SUM(e.poblaci_n_5_16), 3
    ) || ' %' AS cobertura_neta_ponderada
FROM educacion_raw e
JOIN dim_departamento d ON e.c_digo_departamento = d.codigo_departamento
WHERE e.cobertura_neta IS NOT NULL AND e.poblaci_n_5_16 > 0
GROUP BY d.departamento
ORDER BY cobertura_neta_ponderada DESC
LIMIT 20;
"""

df_resultado = pd.read_sql_query(query_cobertura_ponderada_fix, conn)
df_resultado
```

Los resultados muestran que departamentos como Casanare, Sucre y Santander presentan los niveles m√°s altos de cobertura neta educativa, superando el 94%. Esto indica que una proporci√≥n muy alta de la poblaci√≥n en edad escolar (5 a 16 a√±os) en estos territorios est√° accediendo efectivamente al sistema educativo. El an√°lisis tiene en cuenta tanto la cobertura registrada como la poblaci√≥n total en ese grupo etario, lo cual permite comparar de forma m√°s justa entre departamentos de distinto tama√±o. Esta cobertura ponderada refleja no solo eficiencia en matr√≠cula, sino tambi√©n una adecuada planificaci√≥n educativa y atenci√≥n a la demanda escolar real.
